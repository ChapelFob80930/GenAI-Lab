```python
import pandas as pd
import numpy as np
import faiss
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sentence_transformers import SentenceTransformer
from transformers import pipeline
```

```python
nlp = spacy.load("en_core_web_sm")
stop_words = set(stopwords.words('english'))
embedder = SentenceTransformer('all-MiniLM-L6-v2')

generator = pipeline('text2text-generation', model='t5-base', )
```

```python
def preprocess_text(text):
  tokens = word_tokenize(text.lower())
  tokens = [t for t in tokens if t.isalnum() and t not in stop_words]
  return " ".join(tokens)
```

```python
def load_structured_data(path):
  df = pd.read_csv(path)
  docs = []
  for _, row in df.iterrows():
    combined = " ".join(map(str, row.values))
    docs.append(preprocess_text(combined))
  return docs

def load_unstructured_data(path):
  with open(path, 'r') as file:
    content = file.read()
  return [preprocess_text(content)]
```

```python
def build_faiss_index(docs):
  embeddings = embedder.encode(docs, convert_to_numpy=True)
  dimension = embeddings.shape[1]
  index = faiss.IndexFlatL2(dimension)
  index.add(embeddings)
  return index, docs, embeddings
```

```python
def retrieve_docs(query, docs, embeddings, k = 3):
  query_embedding = embedder.encode([query], convert_to_numpy=True)
  D, I = faiss.IndexFlatL2(embeddings.shape[1]).search(query_embedding, k)
  return [docs[i] for i in I[0]]
def generate_answer(query, retrived_docs):
  context = " ".join(retrieve_docs)
  input_text = f"question: {query} context: {context}"
  result = generator(input_text, max_length=100)[0]["generated_text"]
  return result
```
